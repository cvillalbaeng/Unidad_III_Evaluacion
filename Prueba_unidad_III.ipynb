# üõ§Ô∏è Ejercicio 1: Control de Tren (Valor - 5 pts)

### Descripci√≥n
Simula el sistema de control de un tren que debe responder a se√±ales en la v√≠a.  
Las se√±ales pueden ser **"verde"**, **"amarillo"** o **"rojo"**.  
El tren debe ajustar su velocidad base seg√∫n la se√±al actual y la distancia hasta la pr√≥xima estaci√≥n:

- **Se√±al verde**:  
  - Acelera a **100 km/h** si la distancia es mayor a 5 km.  
  - Mantiene **80 km/h** en caso contrario.  

- **Se√±al amarilla**:  
  - Reduce la velocidad a **50 km/h** si la distancia es mayor a 3 km.  
  - De lo contrario, reduce a **30 km/h**.  

- **Se√±al roja**:  
  - Frena hasta **0 km/h** inmediatamente.  

- **Obst√°culo en la v√≠a**:  
  - El tren debe frenar a **0 km/h**, independientemente de la se√±al.  

---

### ‚úÖ Salida esperada
El programa debe devolver la velocidad actual del tren en base a las condiciones dadas.  

**Ejemplo:**
- Entrada: `se√±al="amarillo", distancia=4, obst√°culo=False`  
- Salida: `Velocidad: 50 km/h`

```python
# Aqu√≠ se resuelve el ejercicio

se√±al = input("Indique cual es la se√±al actual (verde, amarilla, roja): ").lower()  # Convierte la respuesta a min√∫sculas
distancia = float(input("Indique la distancia a la estaci√≥n (km): ")) 
obstaculo = int(input("¬øHay obst√°culo en la V√≠a? indique: 1 si hay alg√∫n obst√°culo \n indique 0 si NO hay obst√°culo: ")) # convierte str a entero
obstaculo = bool(obstaculo) # convierte entero a boleano

if obstaculo or se√±al == "roja":
    velocidad_actual = 0
    estado_tren = "Frenado completo por seguridad"

elif se√±al == "amarilla":
    if distancia > 3:
        velocidad_actual = 50
    else:
        velocidad_actual = 30
    estado_tren = "Precauci√≥n: Reduciendo velocidad"

elif se√±al == "verde":
    if distancia > 5:
        velocidad_actual = 100
    else:
        velocidad_actual = 80
    estado_tren = "Acelerando"

else:
    velocidad_actual = 0
    estado_tren = "Error: Se√±al no reconocida"

print(f"\nResultado: {estado_tren}")  # muestra cual es el estado actual del tren
print(f"Velocidad final: {velocidad_actual} km/h")  # muestra cual es la velocidad actual del tren
```

# ‚ò¢Ô∏è Ejercicio 2: Mitigaci√≥n de Fusi√≥n Nuclear (Valor - 5 pts)

### Descripci√≥n
Eres responsable de monitorear un **reactor nuclear**.  
Debes determinar el estado del reactor en base a la **temperatura**, la **presi√≥n** y el **nivel de refrigerante**.

Las reglas son:

- Si la **temperatura > 800 ¬∞C** y la **presi√≥n > 500 bar**, el estado es **"CR√çTICO"**.  
- Si la **temperatura est√° entre 600 ¬∞C y 800 ¬∞C** o la **presi√≥n entre 300 bar y 500 bar**, el estado es **"ALERTA"**.  
- Si el **nivel de refrigerante < 50%**, el estado se degrada un nivel (ejemplo: de "NORMAL" a "ALERTA").  
- En cualquier otro caso, el estado es **"NORMAL"**.  

| Condici√≥n Base | Refrigerante ‚â• 50% | Refrigerante < 50% |
|---------------|-------------------|--------------------|
| NORMAL        | NORMAL            | ALERTA             |
| ALERTA        | ALERTA            | CR√çTICO            |
| CR√çTICO       | CR√çTICO           | CR√çTICO            |
---

### ‚úÖ Salida esperada
El programa debe imprimir el estado del reactor.  

**Ejemplo:**
- Entrada: `temperatura=750, presi√≥n=400, refrigerante=45`  
- Salida: `Estado: CR√çTICO`

```python
# Aqu√≠ se resuelve el ejercicio

temperatura = float(input("Indique cual es la Temperatura actual del Reactor (¬∞C): "))
presion = float(input(" Indique cual es la Presi√≥n actual del Reactor (bar): "))
refrigerante = float(input("Indique cual es el Nivel actual de refrigerante del Reactor (%): "))


if temperatura > 800 and presion > 500:
    estado = "CR√çTICO"
elif (600 <= temperatura <= 800) or (300 <= presion <= 500):
    estado = "ALERTA"
else:
    estado = "NORMAL"

if refrigerante < 50:
    if estado == "NORMAL":
        estado = "ALERTA"
    elif estado == "ALERTA":
        estado = "CR√çTICO"
  
print(f"Estado: {estado}")
```

# üí∞ Ejercicio 3: C√°lculo de Impuestos (Valor - 5 pts)

### Descripci√≥n
Calcula el **impuesto sobre la renta** en base al **ingreso anual**, la **edad** y el **n√∫mero de dependientes**.  

Las reglas son:

- Ingresos **‚â§ $10,000**: no pagan impuestos.  
- Ingresos **entre $10,001 y $50,000**: pagan **10% sobre el exceso** de $10,000.  
- Ingresos **> $50,000**: pagan **20% sobre el exceso de $50,000** m√°s **$4,000**.  
- Mayores de **65 a√±os**: reducci√≥n del **5% en el impuesto total**.  
- Por cada **dependiente**, se reduce **$1,000 del ingreso gravable** antes de calcular el impuesto.  

---

### ‚úÖ Salida esperada
El programa debe calcular y mostrar el impuesto a pagar.  

**Ejemplo:**
- Entrada: `ingreso=60000, edad=70, dependientes=2`  
- Salida: `Impuesto: 5320`  

üìå *Explicaci√≥n:*  
- Ingreso gravable = 60000 - (1000 √ó 2) = **58000**  
- Impuesto base = 4000 + 20% de (58000 - 50000) = **5600**  
- Descuento por edad = 5% de 5600 = **280**  
- Impuesto final = 5600 - 280 = **5320**

```python
# Aqu√≠ se resuelve el ejercicio
ingreso = float(input("Indique cual es la cantidad de su ingreso anual: "))
edad = int(input("Indique cual es su edad: "))
dependientes = int(input("Ingrese el n√∫mero de personas dependientes de usted: "))

ingreso_gravable = ingreso - (dependientes * 1000) # Determina cual es el monto de ingreso gravable, en base al monto inicial de ingreso restandole 1000 por cada persona dependiente que tenga

if ingreso_gravable <= 10000: # si el ingreso gravable es inferior a 1000, no paga impuestos
    impuesto_base = 0

elif 10001 <= ingreso_gravable <= 50000:  # si el ingreso gravable es inferior esta entre 10001 y 50000, resta 1000 y multiplica el restante por 10% para determinar que pagara de impuesto el 10 % solo del excdente de 1000
    impuesto_base = (ingreso_gravable - 10000) * 0.10

else:
    impuesto_base = 4000 + (ingreso_gravable - 50000) * 0.20  # si el ingreso gravable es mayor a 50000, le suma 4000, luego le resta ingreso garvabable los 50000 y a su resultado lo multiplica por el 20% para determinar que pagara de impuesto el 20 % solo del excedente de 50000 + 4000

if edad > 65:
    impuesto_final = impuesto_base * 0.95  # Se resta el 5%
else:
    impuesto_final = impuesto_base

print(f"Impuesto: {impuesto_final}")
```

# üì¶ Ejercicio 4: Sistema de Descuentos en Tienda (Valor - 5 pts)

### Descripci√≥n
Eres responsable de desarrollar la l√≥gica de un sistema de ventas en una tienda.

El programa debe determinar el **descuento aplicado** a un cliente seg√∫n:

- El **tipo de cliente**

Debes utilizar obligatoriamente:

‚úî `match-case` para evaluar el tipo de cliente  
‚úî `if / elif / else` dentro de los casos  
‚úî Al menos un **anidamiento**

---

### Tipos de cliente

- `"regular"`
- `"vip"`
- `"mayorista"`

---

### Reglas

#### 1. Descuento base seg√∫n tipo de cliente (usar `match-case`)

| Tipo de cliente | Descuento base |
|-----------------|----------------|
| regular         | 5%             |
| vip             | 10%            |
| mayorista       | 15%            |

---

#### 2. Ajuste por monto de compra

Dentro de cada tipo de cliente:

- Si el monto es **mayor a 200**, se suma **5% adicional**
- Si el monto es **menor a 50**, el descuento se reduce en **2%**

---

#### 3. Horario especial

Si la compra se realiza entre las **18 y 21 horas**, se suma **3% adicional**

---

#### 4. Descuento m√°ximo

El descuento total **no puede superar el 20%**

---

### Datos de entrada

El programa debe recibir:

- `tipo_cliente`
- `monto_compra`
- `hora_compra`

---

### Tabla de referencia

| Condici√≥n | Resultado |
|-----------|----------|
| Descuento base | Seg√∫n tipo de cliente |

| Monto < 50 | -2% |

---

### ‚úÖ Salida esperada

El programa debe imprimir el descuento final aplicado.
##### Ejemplo
**Entrada:**
`tipo_cliente = "vip"`  
`monto_compra = 220`  
`hora_compra = 19`

**Salida:**
`Descuento final: 18%`

```python
# Aqui se resuelve el ejercicio

tipo_cliente = input("Indique que Tipo de cliente es (regular, vip, mayorista): ").lower()
monto_compra = float(input("Indique cual es el Monto de la compra: "))
hora_compra = int(input("Indique cual es la Hora de compra (0-23 hrs): "))

if tipo_cliente == "regular":  # Define el descuento inicial seg√∫n el tipo de cliente
    descuento = 5
elif tipo_cliente == "vip":
    descuento = 10
elif tipo_cliente == "mayorista":
    descuento = 15
else:
    descuento = 0
 
if monto_compra > 200:     # Ajusta el descuento seg√∫n el monto de la compra
    descuento = descuento + 5
elif monto_compra < 50:
    descuento = descuento - 2


if 18 <= hora_compra <= 21:  # Ajusta el descuento por horario especial (18 a 21 horas)
    descuento = descuento + 3


if descuento > 20:  # Aplica los l√≠mites (M√°ximo 20% y M√≠nimo 0%)
    descuento = 20
if descuento < 0:
    descuento = 0


print(f"El Descuento final es de: {descuento}%")
monto_final = monto_compra - descuento
print(f"Total a pagar: {monto_final}")
```
